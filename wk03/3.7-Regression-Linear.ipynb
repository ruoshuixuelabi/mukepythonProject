```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets # 数据集
from sklearn.linear_model import LinearRegression # 线性回归
from sklearn.model_selection import train_test_split # 数据集拆分
```

```python
# 3.7.1 导入数据、回归vs分类算法
```

```python
data = datasets.load_boston() # 结果集：.data房价影响因素 .target房价
x = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.DataFrame(data.target, columns=['BostonPrice'])
```

```python
# 3.7.2 数据集拆分：train、test
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape ,y_train.shape
```

```python
# 3.7.3 模型训练：LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# 获取系数和截距：y=ax+b -> y=ax1 + bx2 + ..... hx13 + z
lr.coef_ , lr.intercept_
```

```python
# 3.7.4 模型预测：预测值(y_pred) vs 真实值(y_test)
```

```python
y_pred = lr.predict(x_test) # 获取预测值
```

```python
y_pred
```

```python
plt.plot(np.arange(len(y_test)), y_test)
plt.plot(np.arange(len(y_test)), y_pred, color='red')
```

```python
# 3.7.5 模型评估：MSE（均方误差）、数据量调整
```

```python
from sklearn import metrics
metrics.mean_squared_error(y_test,y_pred)
```
