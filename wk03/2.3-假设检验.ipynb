```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as st
```

```python
# 2.3.1 分布图：总体vs样本
```

```python
from scipy.stats import norm

# 设置一下总体样本的概率分布
mean0 = 72
mean1 = 68
std1 = 0.8

s0 = norm.rvs(mean0,1,1000)
s1 = norm.rvs(mean1,std1,100)

sns.distplot(s0)
sns.distplot(s1)
```

```python
# 2.3.2 置信区间和拒绝域：α=0.05，p越小约能拒绝零假设
```

```python
# 置信区间：总体概率为95%
plt.figure(figsize=(10,5))
sns.distplot(norm.rvs(size=1000))
```

```python
# 当我们的类别数据 > 3方差，基本上就是小概率时间，即：可以拒绝原假设
# 拒绝域：> 3std，
```

```python
norm.cdf(2)
```

```python
norm.ppf(0.025) + 72
```

```python
# 2.3.3 简单计算p值（z-score）
```

```python
zscore = (mean1 - mean0) / std1
zscore # 举例期望有多少个方差
```
