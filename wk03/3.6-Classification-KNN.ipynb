```python
from sklearn.datasets import load_iris # 数据集
from sklearn.neighbors import KNeighborsClassifier # KNN估计器
from sklearn.model_selection import train_test_split # 数据集拆分
from sklearn.metrics import classification_report # 分析预测结果
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 3.6.1 加载鸢尾花数据集
```

```python
iris = load_iris()
iris
```

```python
x = iris.data # 4个特征值
y = iris.target # 每个数据样本的所属类别
```

```python
x2d = x[:,:2]
```

```python
sns.set()
sns.scatterplot(x2d[:,0],x2d[:,1],hue=y)
```

```python
# 3.6.2 交叉验证：训练集和测试集
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
x_test.shape
```

```python
# 3.6.3 KNN算法
```

```python
knc = KNeighborsClassifier() # 默认k=5
```

```python
KNeighborsClassifier?
```

```python
knc.fit(x_train,y_train) # 模型训练
```

```python
y_predict = knc.predict(x_test)
```

```python
# 3.6.4 评估KNN的准确性
```

```python
scores = classification_report(y_test,y_predict,
                               target_names=iris.target_names,output_dict=True)
```

```python
import pandas as pd
pd.DataFrame(scores)
```
