```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats
```

### 2.5.1 什么是ks-test

* 一种基于累计分布函数的非参数检验
* KS统计量：D=max|Fn(x)-F0(x)|，即样本所属总体的分布与给定分布之间的距离，越小越相同
* p值：根据D值找到对应类别数据并计算累计概率



```python
# 2.5.2 创建随机变量：近似分布、不同分布
```

```python
size=1000

s1 = stats.norm.rvs(loc=5,scale=5,size=size) # 正态分布
s2 = stats.t.rvs(20,loc=5,scale=5,size=size) # t-分布
s3 = np.random.exponential(size=size) * 5 # 指数分布
```

```python
plt.figure(figsize=(15,5)) # 设置画图大小

for s in [s1,s2,s3]:
    sns.distplot(s) # 绘制直方图
    
plt.legend(labels=[1,2,3]) # 设置标签
```

```python
# 2.5.3 双样本kstest：统计量D、p值
```

```python
stats.ks_2samp(s1,s2)
```

```python
stats.ks_2samp(s1,s3)
```

```python
stats.ks_2samp(s2,s3)
```

```python
# 2.5.3 双样本ttest
```

```python
stats.ttest_ind(s1,s2)
```

```python
stats.ttest_ind(s1,s3)
```

```python
stats.ttest_ind(s2,s3)
```
