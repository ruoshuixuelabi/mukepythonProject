```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as st
```

```python
# 2.2.1 创建随机变量：设置size、设置loc期望值、设置scale方差
```

```python
s1 = st.norm.rvs(loc=0,scale=1,size=1000)
```

```python
sns.distplot(s1)
```

```python
# 2.2.2 不同取值和方差的表现：均值决定中心位、方差决定高矮
```

```python
# 改动方差，观察变化
s1 = st.norm.rvs(loc=0,scale=1,size=1000)
s2 = st.norm.rvs(loc=0,scale=2,size=1000)
s3 = st.norm.rvs(loc=0,scale=3,size=1000)
```

```python
for s in [s1,s2,s3]:
    sns.distplot(s)
plt.legend(labels=[1,2,3])
```

```python
# 改动均值，观察变化
s1 = st.norm.rvs(loc=0,scale=1,size=100)
s2 = st.norm.rvs(loc=1,scale=1,size=100)
s3 = st.norm.rvs(loc=2,scale=1,size=100)
```

```python
for s in [s1,s2,s3]:
    sns.distplot(s)
plt.legend(labels=[1,2,3])
```

```python
# 2.2.3 随机变量的数值特征：期望、方差，样本量越大越准确
```

```python
for s in [s1,s2,s3]:
    print(s.mean(),s.std())
```
